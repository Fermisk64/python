```python
# qualifying for an eye exam
def qualified_for_exam (age, qualifying_condition, time_since_last_exam):
    '''assume (1) age is passed as an int for the age of the person, 
    (2) qualifying_condition is passed as boolean indicating whether qualified for an exam 
    (3) time_since_last exam is passed as an int for the number of months since last exam'''
    
    if 0< age <= 19:
        if time_since_last_exam >= 12:
            return "You are eligible for one eye exam and any minor assessment"
        else:
            return "It has not been more than 12 months for another eye exam"
    elif 20 <= age <= 64:
        if qualifying_condition:   #Since it isn't defined, it automatically considers it as true
            if time_since_last_exam >= 12:
                return "You are eligible for one major eye exam and 2 additional follow-up minor assessment"
            else:
                return "It hasn't been 12 months since your last eye exam"
        else:
            return "You do not have a qualifying eligibla medical condition"
    elif age >= 65:
        if time_since_last_exam >= 18:
            return "You are eligible for one major eye exam or two minor assessments"
        else:
            return "It hasn't been 18 months since your last exam"
    else:
        return "Invalid age input"
```

```python
qualified_for_exam (5, False, 11)
```

```python
qualified_for_exam (30, True, 11)
```

```python
# set operations
rainbow = {'red', 'yellow', 'green', 'blue'}
olympic_flag = {'green', 'yellow', 'black', 'blue'}
```

```python
rainbow.difference(olympic_flag) # it will find values that are in the first but not the second set
```

```python
olympic_flag.difference(rainbow)
```

```python
# symmetric difference: find values that are only in one of the two sets
rainbow.symmetric_difference(olympic_flag)
```

```python
rainbow.intersection(olympic_flag) # find the values that are common in the 2 sets
```

```python
# union: combine the 2 sets to get the unique values in both sets
rainbow.union(olympic_flag)
```
